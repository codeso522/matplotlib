```python
from matplotlib import pyplot as plt

```

```python
plt.style.use("fivethirtyeight")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#how far a part is from radius
explode=[0,0,0,0.1,0]

plt.pie(slices,labels = labels,explode=explode,shadow=True,wedgeprops = {'edgecolor':'black'},startangle=90,autopct='%1.1f%%')
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
